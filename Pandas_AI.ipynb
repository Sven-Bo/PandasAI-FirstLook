## Sample Dataset

```python
import pandas as pd
```

```python
# Sample DataFrame
df = pd.DataFrame({
    "country": ["United States", "United Kingdom", "France", "Germany", "Italy", "Spain", "Canada", "Australia", "Japan", "China"],
    "gdp": [19294482071552, 2891615567872, 2411255037952, 3435817336832, 1745433788416, 1181205135360, 1607402389504, 1490967855104, 4380756541440, 14631844184064],
    "happiness_index": [6.94, 7.16, 6.66, 7.07, 6.38, 6.4, 7.23, 7.22, 5.87, 5.12]
})
df.head()
```

## Pandas

### Which are the 5 happiest countries?

```python
happiest_countries = df.nlargest(5, 'happiness_index')['country'].tolist()
print(f"The 5 happiest countries are: {', '.join(happiest_countries)}")
```

### What is the sum of the GDPs of the 2 unhappiest countries?

```python
unhappiest_gdp_sum = df.nsmallest(2, 'happiness_index')['gdp'].sum()
print(f"The sum of the GDPs of the 2 unhappiest countries is: {unhappiest_gdp_sum:,}")
```

### Plot the histogram of countries showing for each the gpd, using different colors for each bar

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a color palette with as many colors as there are unique countries
colors = sns.color_palette('hsv', len(df['country'].unique()))

df.set_index('country')['gdp'].plot(kind='bar', figsize=(8, 4), color=colors)

plt.title("GDP by Country")
plt.xlabel("Country")
plt.ylabel("GDP")
plt.show()
```

## Pandas AI

> `PandasAI` is designed to be used in conjunction with Pandas. It makes Pandas conversational, allowing you to ask questions about your data and get answers back, in the form of Pandas DataFrames. 
_[Source:https://github.com/gventuri/pandas-ai](https://github.com/gventuri/pandas-ai)_

### Sample Dataset

```python
# Sample DataFrame
df = pd.DataFrame({
    "country": ["United States", "United Kingdom", "France", "Germany", "Italy", "Spain", "Canada", "Australia", "Japan", "China"],
    "gdp": [19294482071552, 2891615567872, 2411255037952, 3435817336832, 1745433788416, 1181205135360, 1607402389504, 1490967855104, 4380756541440, 14631844184064],
    "happiness_index": [6.94, 7.16, 6.66, 7.07, 6.38, 6.4, 7.23, 7.22, 5.87, 5.12]
})
df.head()
```

```python
# Instantiate a LLM
from pandasai import PandasAI
from pandasai.llm.openai import OpenAI
llm = OpenAI(api_token="<YOUR_API_KEY>")
```

```python
pandas_ai = PandasAI(llm, conversational=False)
```

### Which are the 5 happiest countries?

```python
pandas_ai.run(df, prompt='Which are the 5 happiest countries?')
```

### What is the sum of the GDPs of the 2 unhappiest countries?

```python
pandas_ai.run(df, prompt='What is the sum of the GDPs of the 2 unhappiest countries?')
```

### Plot the histogram of countries showing for each the gpd, using different colors for each bar
